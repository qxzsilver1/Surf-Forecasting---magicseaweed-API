```python
import requests
import json
from config import api_key

base_url = "http://magicseaweed.com/api/"

# Rockaway, NY
spot_id = "384"
units = "us"
```

```python
query_url = f"{base_url}{api_key}/forecast/?spot_id={spot_id}&units={units}"
rockaway_swell_response = requests.get(query_url)
rockaway_swell_json = rockaway_swell_response.json()
```

```python
print(json.dumps(rockaway_swell_json, indent=4, sort_keys=True))
```
