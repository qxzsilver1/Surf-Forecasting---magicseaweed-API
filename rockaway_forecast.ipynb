```python
import requests
import json
from config import api_key
from datetime import datetime
from collections import namedtuple

base_url = "http://magicseaweed.com/api/"

# Rockaway, NY
spot_id = "384"
units = "us"
```

```python
query_url = f"{base_url}{api_key}/forecast/?spot_id={spot_id}&units={units}"
rockaway_swell_response = requests.get(query_url)
rockaway_swell_json = rockaway_swell_response.json()
```

```python
print(json.dumps(rockaway_swell_json, indent=4, sort_keys=True))
```

```python
dt_list = []

for swell_info in rockaway_swell_json:
    dt_obj = datetime.fromtimestamp(swell_info["timestamp"])
    dt_list.append(dt_obj)
```

```python
dt_list
```

```python
def _json_object_hook(d): return namedtuple('X', d.keys())(*d.values())
def json2obj(data): return json.loads(data, object_hook=_json_object_hook)

json_list = []
for swell_info in rockaway_swell_json:
    x = json2obj(str(swell_info).replace("\'","\"")) # namedtuple needs double quotes instead of single quotes - replace these in JSON string
    json_list.append(x)
```
